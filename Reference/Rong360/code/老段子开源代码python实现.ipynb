# 预测：

```python
预测表 = 测试表
```

测试信用卡账单表#414894行

```python
测试信用卡账单表.drop_duplicates()
```

```python
t1=测试信用卡账单表[(测试信用卡账单表['时间']>测试信用卡账单表['放款时间'])].groupby("用户标识",as_index=False)
t2=测试信用卡账单表[(测试信用卡账单表['时间']>测试信用卡账单表['放款时间']+1)].groupby("用户标识",as_index=False)
t3=测试信用卡账单表[(测试信用卡账单表['时间']>测试信用卡账单表['放款时间']+2)].groupby("用户标识",as_index=False)
```

```python
t1=t1['时间'].agg({'t1' : 'count'})
t2=t2['时间'].agg({'t2' : 'count'})
t3=t3['时间'].agg({'t3' : 'count'})
```

```python
预测表=pd.merge(预测表,t1,how='left',on = "用户标识")
预测表=pd.merge(预测表,t2,how='left',on = "用户标识")
预测表=pd.merge(预测表,t3,how='left',on = "用户标识")
```

```python
预测表=预测表.fillna(0)
```

```python
预测表['预测']=-0.4*np.log(预测表['t1']+1)-0.4*np.log(预测表['t2']+1)-0.4*np.log(预测表['t3']+1)
```

```python
预测表['预测']=预测表['预测']-min(预测表['预测'])
预测表['预测']=预测表['预测']/max(预测表['预测'])
```

```python
output=预测表[['用户标识','预测']].rename(index=str, columns={"用户标识": "userid", "预测": "probability"})
```

```python
output.to_csv("output/rule_v2.csv",index=None,encoding='utf-8')
```
